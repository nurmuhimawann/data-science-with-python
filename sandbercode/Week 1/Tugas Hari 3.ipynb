<h2>Soal 1: Comparison Operator</h2>

- Berikan contoh comparison dari tipe data string
- Berikan contoh comparison dari tipe data boolean
- Berikan contoh comparison dari tipe data integer

```python
# string comparison
string = "Nur Muhammad Himawan"
print("comparison tipe data string: {0}".format ("Nur Muhammad Himawan" == string))

# bool comparison
print("comparison tipe data boolean: {0}".format (True != False))

# int comparison
print("comparison tipe data integer: {0}".format (3 == 3))
```

<h2>Soal 2: Boolean Comparison</h2>

- Berikan contoh gabungan multiple comparison menggunkan Boolean Comparisin 'and'
- Berikan contoh gabungan multiple comparison menggunkan Boolean Comparisin 'or'
- Berikan contoh gabungan multiple comparison menggunkan Boolean Comparisin 'not'

```python
# comparison and
# akan mereturn nilai True apabila kedua kondisi terpenuhi
x = 1
print(x > 0 and x < 10)

# comparison or
# akan mereturn nilai True apabila salah satu dari kedua kondisi terpenuhi
x = 2
print(x > 5 or x < 10)

# comparison not
# comparison not digunakan untuk membalikkan nilai hasil.
# Pada contoh menghasilkan result 'True', lalu di reverse oleh not menjadi 'False'
x = 3
print(not(x > 0))
```

<h2>Soal 4: Comparison Operator dengan fungsi</h2>

Buatlah sebuah fungsi yang menerima satu argument bertipe data numeric dan menghasilkan sebuah return sebagai berikut :
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan ganjil
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 2 sampai 5 (2 dan 5 termasuk)
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 6 sampai 20 (6 dan 20 termasuk)
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan lebih besari dari 20

```python
# type data is numeric
x = 10

# Buat fungsi
def cek_nilai(value):
    if value % 2 == 1:
        return ("Aneh")
    elif (value % 2 == 0) and (2 <= value <= 5):
        return ("Tidak Aneh")
    elif (value % 2 == 0) and (6 <= value <= 20):
        return ("Aneh")
    elif (value % 2 == 0) and (value > 20):
        return ("Tidak Aneh")

cek_nilai(x)
```

<h2>Soal 5: While Loop dan For Loop</h2>

- Apa perbedaan while loop dan for loop?
- Berikan contoh sederhana cara menggunakan while loop dan for loop

```python
"""
Apa perbedaan while-loop dan for-loop?
Jawab:
While-loop dan for-loop adalah blok kode yang digunakan untuk menjalankan perintah perulangan (looping). menurut saya, 
perbedaannya hanya terdapat pada segi penggunaannya saja, while-loop digunakan ketika jumlah perulangannya belum ditentukan
(uncountable), while-loop memiliki syarat tertentu untuk perulangannya.
Sebaliknya, for-loop digunakan dalam perulangan yang sudah diketahui berapa banyak jumlah perulangannya (countable).
"""

# Tulis Kode
# for-loop untuk countable
# disini kita sudah tahu berapa banyak perulangan yang kita inginkan dengan memanfaatkan 'range'
for i in range(10):
    print(i, end= ' ')

# while-loop untuk uncountable, terdapat syarat
# disini kita tidak tahu berapa banyak perulangan yang dilakukan,
# yang kita tahu ketika variabel 'value' nya bernilai 'False' maka perulangan akan berhenti
value = 1
while value <= 20:
    print("halo, it's me")
    value += 1
```

<h2>Soal 6: While Loop</h2>
    
Buatlah sebuah code while loop sebagai berikut :
- Buatlah suatu fungsi yang merima satu input nilai numeric
- Dalam fungsi, Buat sebuah while-loop dengan melakukan comparasi terhadap inputan tersebut, dimana looping/iterasi terus berjalan apabila nilai inputan tidak sama dengan 0
- Dalam setiap iterasi update nilai variable inputan itu, jika input lebih dari 0, maka kurangi variabel tersebut sebesar 1, selain itu tambahkan 1
- print nilai variable input dalam setiap awal iterasi

```python
a = 10

def fungsi_while(value):
    # apabila value bernilai positif
    if value > 0:
        while value != 0:
            print(value)
            value -= 1
    
    # apabila value bernilai negatif
    elif value < 0:
        while value != 0:
            print(value)
            value += 1

    # apabila value bernilai 0
    else:
        print(value)

fungsi_while(a)
```
