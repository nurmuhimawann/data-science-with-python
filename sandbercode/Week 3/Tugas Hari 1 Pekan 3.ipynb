<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
plt.style.use('seaborn')
y = np.sin(x/2)
y2 = np.cos(x)
y3 = np.cos(2*x)
plt.plot(y)
plt.plot(y2)
plt.plot(y3)
plt.show()
```

```python
#menyimpan hasil figure
fig, ax = plt.subplots()
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
plt.style.use('seaborn')
y = np.sin(x/2)
y2 = np.cos(x)
y3 = np.cos(2*x)
plt.plot(y)
plt.plot(y2)
plt.plot(y3)
plt.show()
fig.savefig('coba.png')

#menampilkan figure yang disimpan
from IPython.display import Image
Image('coba.png')
```
