```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(20,8), nrows=1, ncols=3, sharex=True, sharey=True)
# grafik tan(x)
ax[0].plot(x, tan, label='tanx', color='blue')
ax[0].set_title('Grafik tanx')
# grafik sin(x)
ax[1].plot(x, sin, label='sinx', color='red')
ax[1].set_title('Grafik sinx')
# grafik cos(x)
ax[2].plot(x, cos, label='cosx', color='green')
ax[2].set_title('Grafik cosx')
# set label
fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(15,10))
ax.set_title('Grafik CosX')
ax.plot(x, y)
# inset 1
ins1 = ax.inset_axes([0.8, 0.8, 0.3, 0.3])
ins1.plot(x, y2, color='black')
ins1.plot(x, y, color='blue')
ins1.set_title('Grafik CosX^4')
# inset 2
ins2 = ax.inset_axes([0.8, -0.1, 0.3, 0.3])
ins2.plot(x, y3, color='green')
ins2.plot(x, y, color='blue')
ins2.set_title('Grafik CosX^3')
# inset 3
ins3 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins3.plot(x, y4, color='red')
ins3.plot(x, y, color='blue')
ins3.set_title('Grafik CosX^2')
# inset 4
ins4 = ax.inset_axes([-0.1, 0.8, 0.3, 0.3])
ins4.plot(x, y5, color='y')
ins4.plot(x, y, color='blue')
ins4.set_title('Grafik CosX^5')

plt.show()
```
