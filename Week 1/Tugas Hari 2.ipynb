Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

//isi jawaban text disini


```python
# Apa itu built-in Function
"""
Jawab:
fungsi yang sudah ada di python atau built-in (bawaan) tanpa kita harus menginstall apapun untuk menjalankannya.
"""

# Sebutkan 3 built-in Function di python!
"""
Jawab:
Diantaranya ada str(), min(), dan len().
"""

# Contoh penggunaan built in function
# 1. str() digunakan untuk mengembalikan object string
x = 8
x = str(x)
print(x)
print(type(x))

# 2. min() digunakan untuk mendapatkan nilai minimal
a = [100, 5, 8, 3, 87, 99]
print(a)
print(f"Nilai minimum dari array diatas adalah {min(a)}")

# 3. len() digunakan untuk menghitung panjang karakter
nama_saya = "Nur Muhammad Himawan"
print(nama_saya)
print(f"Panjang karakter string diatas adalah {len(nama_saya)} karakter")
```

<h2>Soal 4: Menggunkan Method String</h2>

Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
kalimat = "Corona cepat selesai"

# gunakan method untuk mengubah nilai kalimat menjadi uppercase semua kemudian tampilkan hasilnya
print(kalimat.upper())

# gunakan method untuk menghitung berapa huruf e di dalam kalimat
print(kalimat.count("e"))
```

<h2>Soal 5: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]
def mean_list(inp_list):
    # isikan code
    return (sum(inp_list) / len(inp_list))
    
print(mean_list(obj_list))
```

Expected Output:

13.9

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

def kali_list(obj_list, obj_penambah):
    # isikan kode
    return obj_list + obj_penambah
    
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
